```python
import pandas as pd
dataset = pd.read_csv('Toyota.csv', header= 0,
                        encoding= 'unicode_escape')
```

```python
dataset
```

```python
data=dataset[["Price","Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]]
```

```python
data
```

```python
import statsmodels.formula.api as smf 

model1=smf.ols('Price~Age_08_04+KM++HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data).fit()
```

```python
model1.summary()
```

# Checking with each column that has high pvalue.

```python
model1_doors=smf.ols('Price~Doors',data=data).fit()
```

```python
model1_doors.summary()
```

```python
model1_cc=smf.ols('Price~cc',data=data).fit()
```

```python
model1_cc.summary()
```

```python
model1_cc_and_Doors=smf.ols('Price~Doors+cc',data=data).fit()
```

```python
model1_cc_and_Doors.summary()
```

# Creating model without taking cc and doors column as they have high pvalues>0.05

```python
model1_without_cc_and_Doors=smf.ols('Price~Age_08_04+KM++HP+Gears+Quarterly_Tax+Weight',data=data).fit()
```

```python
model1_without_cc_and_Doors.summary()
```

```python
data.corr()
```

```python
rsq_Age = smf.ols('Age_08_04~KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data).fit().rsquared  
vif_Age = 1/(1-rsq_Age) 

rsq_KM = smf.ols('KM~Age_08_04+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data).fit().rsquared  
vif_KM = 1/(1-rsq_KM)

rsq_HP = smf.ols('HP~KM+Age_08_04+cc+Doors+Gears+Quarterly_Tax+Weight',data=data).fit().rsquared  
vif_HP = 1/(1-rsq_HP)

rsq_cc = smf.ols('cc~KM+Age_08_04+Age_08_04+Doors+Gears+Quarterly_Tax+Weight',data=data).fit().rsquared  
vif_cc = 1/(1-rsq_cc)

rsq_Doors = smf.ols('Doors~KM+Age_08_04+cc+Age_08_04+Gears+Quarterly_Tax+Weight',data=data).fit().rsquared  
vif_Doors = 1/(1-rsq_Doors)

rsq_Gears = smf.ols('Gears~KM+Age_08_04+cc+Doors+Age_08_04+Quarterly_Tax+Weight',data=data).fit().rsquared  
vif_Gears = 1/(1-rsq_Gears)


rsq_Quarterly_Tax = smf.ols('Quarterly_Tax~KM+Age_08_04+cc+Doors+Gears+Age_08_04+Weight',data=data).fit().rsquared  
vif_Quarterly_Tax = 1/(1-rsq_Quarterly_Tax)

rsq_Weight = smf.ols('Weight~KM+Age_08_04+cc+Doors+Age_08_04+Quarterly_Tax+Age_08_04',data=data).fit().rsquared  
vif_Weight = 1/(1-rsq_Weight)


# Storing vif values in a data frame
d1 = {'Variables':['Age','KM','HP','cc','Doors','Gears','Quarterly Tax','Weight'],'VIF':[vif_Age,vif_KM,vif_HP,vif_cc,vif_Doors,vif_Gears,vif_Quarterly_Tax,vif_Weight]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

```python
pred1 = model1.predict(data.iloc[:,1:])
pred1
```

```python
pred2 = model1_cc.predict(data.iloc[:,1:])
pred3 = model1_doors.predict(data.iloc[:,1:])
pred4 = model1_cc_and_Doors.predict(data.iloc[:,1:])
pred5 = model1_without_cc_and_Doors.predict(data.iloc[:,1:])
```

```python
import sklearn.metrics
import math
```

```python
mse1 = sklearn.metrics.mean_squared_error(data.iloc[:,0],pred1)
rmse1 = math.sqrt(mse1)

mse2 = sklearn.metrics.mean_squared_error(data.iloc[:,0],pred2)
rmse2 = math.sqrt(mse2)

mse3 = sklearn.metrics.mean_squared_error(data.iloc[:,0],pred3)
rmse3 = math.sqrt(mse3)

mse4 = sklearn.metrics.mean_squared_error(data.iloc[:,0],pred4)
rmse4 = math.sqrt(mse4)

mse5 = sklearn.metrics.mean_squared_error(data.iloc[:,0],pred5)
rmse5 = math.sqrt(mse5)

```
