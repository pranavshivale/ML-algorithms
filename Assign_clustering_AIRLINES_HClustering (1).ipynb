```python
data=pd.read_excel("airline_updated.xlsx")
```

```python
data.corr()
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
airlines = pd.DataFrame(trans.fit_transform(data.iloc[:,1:]))
airlines 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
p = np.array(airlines) 
z = linkage(p, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(p) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
airlines['clust']=cluster_labels 
airlines
```

```python
airlines.iloc[:,1:].groupby(airlines.clust).mean()
```

# Changing distance measure

```python
p = np.array(airlines) 
z = linkage(p, method="complete",metric="Maximum")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
)
plt.show()
```
