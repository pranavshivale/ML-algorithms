```python
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv("SalaryData_Train.csv")
```

```python
train_df=pd.get_dummies(train_data,drop_first=True)
```

```python
train_data.head()
```

```python
train_df
```

```python
X=train_df.iloc[:,:-1]
y=train_df.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
X_df=pd.DataFrame(X)
scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(X_df) 
X_df.loc[:,:] = scaled_values
```

```python
X_df
```

```python
y_df=pd.DataFrame(y)
```

```python
y_df.head()
```

```python
y_df.rename(columns = {'Salary_ >50K':'salary'}, inplace = True) 
```

```python
y_df.head()
```

```python
final_df = pd.concat([X_df,y_df], join = 'outer', axis = 1) 
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

# Now working on test data file...

```python
test_data=pd.read_csv("SalaryData_Test.csv")
```

```python
test_df=pd.get_dummies(train_data,drop_first=True)
```

```python
test_X=test_df.iloc[:,:-1]
test_y=test_df.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
test_X_df=pd.DataFrame(test_X)
scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(test_X_df) 
test_X_df.loc[:,:] = scaled_values
```

```python
test_X_df.head()
```

```python
test_y_df=pd.DataFrame(test_y)
```

```python
test_y_df.rename(columns = {'Salary_ >50K':'salary'}, inplace = True) 
```

```python
test_y_df.head()
```

```python
test_final_df = pd.concat([test_X_df,test_y_df], join = 'outer', axis = 1) 
```

```python
test_final_df.head()
```

```python
test_final_df.isnull().sum()
```

# Now training the model

```python
train_X = final_df.iloc[:,:-1]
train_y = final_df.iloc[:,-1]
test_X  = test_final_df.iloc[:,:-1]
test_y  = test_final_df.iloc[:,-1]
```

```python
from sklearn.svm import SVC

model_linear = SVC(kernel = "linear")
model_linear.fit(train_X,train_y)
pred_test_linear = model_linear.predict(test_X)

np.mean(pred_test_linear==test_y)
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly")
model_poly.fit(train_X,train_y)
pred_test_poly = model_poly.predict(test_X)
np.mean(pred_test_poly==test_y)
```

```python

# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(train_X,train_y)
pred_test_rbf = model_rbf.predict(test_X)

np.mean(pred_test_rbf==test_y)
```

# So we can go with Polynomial kernel
