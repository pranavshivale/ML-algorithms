```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
data1['State'] = ct.fit_transform(data1['State'])
```

```python
#Complete Iris dataset
label_encoder = preprocessing.LabelEncoder()
data1['State']= label_encoder.fit_transform(data1['State']) 
```

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Profit~RD+Administration+Marketing+State',data=data1).fit()
```

```python
model.summary()
```

```python
#Build model
import statsmodels.formula.api as smf 
model_Administration = smf.ols('Profit~Administration',data=data1).fit()
```

```python
model_Administration.summary()
```

```python
#Build model
import statsmodels.formula.api as smf 
model_marketing = smf.ols('Profit~Marketing',data=data1).fit()
```

```python
model_marketing.summary()
```

```python
#Build model
import statsmodels.formula.api as smf 
model_state = smf.ols('Profit~State',data=data1).fit()
```

```python
model_state.summary()
```

```python
#Build model
import statsmodels.formula.api as smf 
model_final = smf.ols('Profit~RD+Marketing',data=data1).fit()
```

```python
model_final.summary()
```

```python
data={"Model":pd.Series(["with_all","with_admin","with_state","with_marketing","final_model"]),
     "RSquare":pd.Series([model.rsquared,model_Administration.rsquared,model_state.rsquared,model_marketing.rsquared,model_final.rsquared])}
table=pd.DataFrame(data)
table
```
