```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("book.csv",encoding='latin-1')
```

```python
data=data.iloc[:,1:4]
```

```python
data
```

```python
data=data.rename(columns={"User.ID": "userid","Book.Title":"title","Book.Rating":"rating"})
```

```python
data.head()
```

```python
#number of unique users in the dataset
len(data.userid.unique())
```

```python
len(data.title.unique())
```

```python
data.drop_duplicates(subset ="userid", 
                     keep = False, inplace = True) 
```

```python
len(data.userid.unique())
```

```python
user_book_df = data.pivot(index='userid',
                                 columns = 'title',
                                 values = 'rating').reset_index(drop=True)
```

```python
user_book_df.iloc[500:600,500:600]
```

```python
user_book_df.index = data.userid.unique()
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
user_book_df
```

```python
#IBCF
#Fetching ratings
Married_user_rating = user_book_df['Married In Haste (Intrigue , No 481)']  
```

```python
Married_user_rating
```

# Here,Finding books similar to "Married In Haste (Intrigue , No 481)" using corrwith function.


```python
#Finding the correlation with different books
similar_to_Married = user_book_df.corrwith(Married_user_rating) 
```

```python
similar_to_Married.head(10) 
```

```python
corr_Married = pd.DataFrame(similar_to_Married, columns=['Correlation'])
corr_Married.dropna(inplace=True)
corr_Married.head()
```

```python
corr_Married[corr_Married['Correlation'] > -0.5].sort_values(by='Correlation', ascending=False).head(5)  
```
