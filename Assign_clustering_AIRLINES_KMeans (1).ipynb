```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("airline_updated.xlsx")
```

```python
data
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(data)
```

```python
for i in y_kmeans:
    print(i)
```

```python
data1=np.array(data)
```

```python
plt.scatter(data1[y_kmeans == 0],data1[y_kmeans==0],s=100,c='red',label='cluster 0')
plt.scatter(data1[y_kmeans == 1],data1[y_kmeans==1],s=100,c='orange',label='cluster 1')
plt.scatter(data1[y_kmeans == 2],data1[y_kmeans==2],s=100,c='blue',label='cluster 2')
plt.scatter(data1[y_kmeans == 3],data1[y_kmeans==3],s=100,c='green',label='cluster 3')
#plt.scatter(kmeans.cluster_centers_[:,0:12],kmeans.cluster_centers_[:,0:12],s=300,c='yellow',label='Centroids')
plt.legend()
plt.show()
```
